```python
import numpy as np
import matplotlib.pyplot as plt
from util.vector_util import *
from util.color_util import *
import matplotlib_inline

matplotlib_inline.backend_inline.set_matplotlib_formats('retina')


class Sphere:
    def __init__(self, center, radius, color, specular, reflective):
        self.center = center
        self.radius = radius
        self.color = color
        self.specular = specular
        self.reflective = reflective

class Lighting:
    def __init__(self, type, intensity, direction=None):
        self.type = type
        self.intensity = intensity
        self.direction = direction

        
viewport_size = 1
d = 1
Cw = 600
Ch = 600
Vw = 1
Vh = 1
inf = 10**9
recurrionDepth = 3
O = (0, 0, 0)
BACKGROUND_COLOR = (0, 0, 0)
```

```python
def CanvasToViewport(x, y):
    return (x*Vw/Cw, y*Vh/Ch, d)
```

```python
def IntersectRaySphere(O, D, sphere): 
    r = sphere.radius
    CO = [O[0] - sphere.center[0], O[1] - sphere.center[1], O[2] - sphere.center[2]]
    a = dot(D, D)
    b = 2 * dot(CO, D)
    c = dot(CO, CO) - r*r
    discriminant = b*b - 4*a*c
    if discriminant < 0:
        return None, None
    t1 = (-b + math.sqrt(discriminant)) / (2*a)
    t2 = (-b - math.sqrt(discriminant)) / (2*a)
    return t1, t2
```

```python
def reflectedRay(N, R):
    return minus(multiplication(N, 2 * dot(N, R)), R)
```

```python
def computeLighting(P, N, lightingObj, specular, spheresObj):
    I = 0.0
    for light in lightingObj:
        if light.type == 'ambient':
            I += light.intensity
        else:
            L = light.direction
            t_max = 10**9
            if light.type == 'point':
                L = minus(light.direction, P)
                t_max = 1
            closest_sphere, _ = findClosestIntersection(P, L, 0.001, t_max, spheresObj)
            if closest_sphere != None:
                continue
            dot_product_L = dot(N, L)
            if dot_product_L >= 0:
                I += light.intensity * dot_product_L/(magnitude(N) * magnitude(L))
            R = reflectedRay(N, L)
            dot_product_R = dot(R, N)
            if dot_product_R > 0:
                I += light.intensity * ((dot_product_R/(magnitude(N) * magnitude(R)))**specular)   
    return I
```

```python
def findClosestIntersection(O, D, t_min, t_max, spheresObj):
    closest_t = 10**9
    closest_sphere = None
    for sphere in spheresObj:
        t1, t2 = IntersectRaySphere(O, D, sphere)
        if t1 != None and t1 >= t_min and t1 <=  t_max and t1 < closest_t:
            closest_t = t1
            closest_sphere = sphere
        
        if t2 != None and t2 >= t_min and t2 <=  t_max and t2 < closest_t :
            closest_t = t2
            closest_sphere = sphere
    return closest_sphere, closest_t
```

```python
def TraceRay(O, D, t_min, t_max, spheresObj, lightingObj, recurrionDepth):
    closest_sphere, closest_t = findClosestIntersection(O, D, t_min, t_max, spheresObj)
    if closest_sphere == None :
        return BACKGROUND_COLOR
    P = addition(O, multiplication(D, closest_t))
    N = unit_vector(minus(P, closest_sphere.center))
    I = computeLighting(P, N, lightingObj, closest_sphere.specular, spheresObj)
    local_color = intensity_multiplier(closest_sphere.color, I)

    r = closest_sphere.reflective
    if recurrionDepth <= 0 or r <= 0.0:
        return local_color
    reflected_ray = reflectedRay(N, multiplication(D, -1))
    reflected_color = TraceRay(P, reflected_ray, 0.001, inf, spheresObj, lightingObj, recurrionDepth - 1)
    return color_add(intensity_multiplier(local_color, (1 - r)), intensity_multiplier(reflected_color, r))
```

```python
def putPixel(x, y, color, matrix):
    Sx = Cw//2 + x
    Sy = Ch//2 - y
    if Sx >= Cw or Sy >= Ch:
        return
    matrix[Sy][Sx] = color #y is the rows and x is the columns
```

```python
def switch_lighting(case):
    if case[0] == 'ambient':
        return Lighting(case[0], case[1])
    elif case[0] == 'point':
        return Lighting(case[0], case[1], case[2])
    elif case[0] == 'directional':
        return Lighting(case[0], case[1], case[2])

def render():
    spheres =  [[(0, -1, 3), 1, (255, 0, 0), 1000, 0.2], 
                [(2, 0, 4), 1 ,(0, 0, 255), 1000, 0.3], 
                [(-2, 0, 4), 1 , (0, 255, 0), 100, 0.4],
                [(0, -5001, 0), 5000 , (255, 255, 0), 1000, 0.5]]
    spheresObj = []
    matrix = [[BACKGROUND_COLOR for i in range(Cw)] for j in range(Ch)]
    for sphere in spheres:
        spheresObj.append(Sphere(sphere[0], sphere[1], sphere[2], sphere[3], sphere[4]))
    lightings =  [["ambient", 0.2],
                  ["point", 0.6, (2, 1, 0)],
                  ["directional", 0.2, (1, 4, 4)]]
    lightingObjs = []
    for i in lightings:
        lightingObjs.append(switch_lighting(i))
        
    for x in range(-Cw//2, Cw//2):
        for y in range(-Ch//2, Ch//2):
            D = CanvasToViewport(x, y)
            color = TraceRay(O, D, 1, 10**9, spheresObj, lightingObjs, recurrionDepth)
            putPixel(x, y, color, matrix)
    return matrix
```

```python
img = np.array(render())
plt.axis("off")
plt.imshow(img)
```
